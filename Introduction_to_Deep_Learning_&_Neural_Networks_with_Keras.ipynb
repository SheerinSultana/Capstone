```
import pandas as pd
import keras 
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import statistics as stat
```

```
data= pd.read_csv('concrete_data.csv')
data_columns =data.columns
predictors = data[data_columns[data_columns != 'Strength']] 
target = data['Strength']
n_cols=predictors.shape[1]
```

**Part A**


```
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```
error_list=[]
for i in range(50):
  x_train,x_test,y_train,y_test=train_test_split(predictors,target,train_size=0.7,random_state=2)
  model.fit(x_train, y_train, validation_split=0.3, epochs=50, verbose=2)
  y_pred=model.predict(x_test)
  error=mean_squared_error(y_test,y_pred)
  error_list.append(error)
```

```
mean_a=stat.mean(error_list)
sd_a=stat.stdev(error_list)
print('mean : ',mean_a)
print('standard deviation:',sd_a)
```

**Part B**

```
predictors_norm = (predictors - predictors.mean()) / predictors.std()
```

```
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```

error_list=[]
for i in range(50):
  x_train,x_test,y_train,y_test=train_test_split(predictors_norm,target,train_size=0.7,random_state=2)
  model.fit(x_train, y_train, validation_split=0.3, epochs=50, verbose=2)
  y_pred=model.predict(x_test)
  error=mean_squared_error(y_test,y_pred)
  error_list.append(error)
```

```
mean_b=stat.mean(error_list)
sd_b=stat.stdev(error_list)
print('mean : ',mean_b)
print('standard deviation:',sd_b)
print('mean of the mean squared errors compare to that from Step A:',mean_a-mean_b)
```
